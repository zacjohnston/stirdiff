```python
import numpy as np
import pandas as pd
```

## Load 3d profiles

```python
data = {}

# columns = ['radius', 'dens', 'pres', 
#            'KErad', 'KEtht', 'KEphi', 
#            'Rrr', 'Rtt', 'Rpp', 'cs2', 'vesc2']

columns = ['radius', 'dens', 'pres', 'Rrr',]

for i in range(120, 130):
    data[i] = pd.read_csv(f'3d/mesa20_v_LR_hdf5_plt_cnt_0{i}.dat', 
                          skiprows=2, 
                          usecols=[0, 1, 2, 6],
                          delim_whitespace=True,
                          header=None,
                          names=columns)
data[120]
```

## Get averages

```python
avg = data[120].copy()
avg *= 0

for i, table in data.items():
    avg += table
    
avg *= 1 / len(data)
```

## Get vcon

```python
Rrr = np.array(avg['Rrr'])
Rrr[Rrr < 0] = 0  # set floor

avg['vcon'] = np.sqrt(Rrr)
avg.to_csv('3d/mesa20_v_LR_avg.dat', index=False)

avg
```

## Integrated turb kinetic energy

```python
tot_turb_kin = np.trapz(4*np.pi*avg['radius']**2 * avg['dens'] * avg['vcon']**2, 
                        x=avg['radius'])

tot_turb_kin
```
